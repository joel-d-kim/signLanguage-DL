```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.decomposition import PCA, FastICA
from sklearn.manifold import TSNE
```

```python
file_dir = './Data//sign_mnist_train//sign_mnist_train.csv'
df = pd.read_csv(file_dir)
df_values = df.values
X = df_values[:, 1:]
y = np.array(df_values[:, 0])
```

```python
X_scaled = X/255.0
```

```python
pca = FastICA(n_components=100)
X_pca = pca.fit_transform(X_scaled) 
```

```python
tsne = TSNE(n_components=2, verbose=1, perplexity=50, n_iter=1000, learning_rate=200)
```

```python
tsne_results = tsne.fit_transform(X_pca)
```

```python
from scipy.spatial import ConvexHull

plt.figure(figsize=(12,12))
for i in range(26):
    inds = y == i
    if inds.sum() == 0:
        continue
    plt.scatter(tsne_results[inds, 0], 
                tsne_results[inds, 1], 
                color=(np.random.rand(),np.random.rand(),np.random.rand()), 
                label=i)
    points = tsne_results[inds]
    hull = ConvexHull(points)
    plt.plot(points[hull.vertices,0], points[hull.vertices,1], lw=2)
    plt.plot(points[hull.vertices[0],0], points[hull.vertices[0],1])
plt.legend()
```

```python
plt.figure(figsize=(12,12))
for i in range(26):
    inds = y == i
    plt.scatter(tsne_results[inds, 0], tsne_results[inds, 1], color=(np.random.rand(),np.random.rand(),np.random.rand()), label=i)
plt.legend()
```
